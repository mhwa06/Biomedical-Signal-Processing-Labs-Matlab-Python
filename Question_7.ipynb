```python
import numpy as np
import scipy.io as scio
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt

# -----------------------------
# 1) Load data
# -----------------------------
# Change paths if needed
Rva1 = scio.loadmat("Rva1.mat")["Rva1"]
Rva2 = scio.loadmat("Rva2.mat")["Rva2"]
Rva3 = scio.loadmat("Rva3.mat")["Rva3"]

Ra1  = scio.loadmat("Ra1.mat")["Ra1"]
Ra2  = scio.loadmat("Ra2.mat")["Ra2"]
Ra3  = scio.loadmat("Ra3.mat")["Ra3"]

# Merge into (12, 15000, 75)
Xva = np.concatenate([Rva1, Rva2, Rva3], axis=2)
Xa  = np.concatenate([Ra1,  Ra2,  Ra3],  axis=2)

fs = 256

# Use V1..V6 => Matlab 7:12 => Python indices 6:12
X_in  = Xva[6:12, :, :]   # (6, 15000, 75)
Y_out = Xa[6:12,  :, :]   # (6, 15000, 75)

# Convert to (subjects, time, channels)
X_in  = np.transpose(X_in,  (2, 1, 0)).astype(np.float32)  # (75,15000,6)
Y_out = np.transpose(Y_out, (2, 1, 0)).astype(np.float32)  # (75,15000,6)

print("X_in:", X_in.shape, "Y_out:", Y_out.shape)

# -----------------------------
# 2) Windowing (creates many samples)
# -----------------------------
WIN = 1024
STRIDE = 256

def make_windows(X, Y, win=WIN, stride=STRIDE):
    xs, ys = [], []
    for s in range(X.shape[0]):
        x = X[s]
        y = Y[s]
        for start in range(0, x.shape[0] - win + 1, stride):
            xs.append(x[start:start+win, :])
            ys.append(y[start:start+win, :])
    return np.stack(xs), np.stack(ys)

Xw, Yw = make_windows(X_in, Y_out)
print("Windows:", Xw.shape, Yw.shape)  # (~75*(15000-1024)/256 ≈ thousands windows)

# -----------------------------
# 3) Normalize (per-window)
# -----------------------------
# Normalize input and target together using input stats (simple + stable)
mu = Xw.mean(axis=1, keepdims=True)
sd = Xw.std(axis=1, keepdims=True) + 1e-6
Xw_n = (Xw - mu) / sd
Yw_n = (Yw - mu) / sd

# -----------------------------
# 4) Train/val split
# -----------------------------
n = Xw_n.shape[0]
perm = np.random.permutation(n)
ntr = int(0.85 * n)
tr_idx, va_idx = perm[:ntr], perm[ntr:]

Xtr, Ytr = Xw_n[tr_idx], Yw_n[tr_idx]
Xva, Yva = Xw_n[va_idx], Yw_n[va_idx]

print("Train:", Xtr.shape, "Val:", Xva.shape)

# -----------------------------
# 5) Model: Residual 1D CNN (denoiser)
# -----------------------------
def build_residual_denoiser(win=WIN, ch=6):
    inp = layers.Input(shape=(win, ch))
    x = inp

    # Feature extractor
    for filters in [32, 32, 64, 64]:
        x = layers.Conv1D(filters, 9, padding="same")(x)
        x = layers.BatchNormalization()(x)
        x = layers.Activation("relu")(x)

    # Predict residual noise / vent component-like correction
    res = layers.Conv1D(ch, 9, padding="same")(x)

    # Residual output
    out = layers.Add()([inp, res])
    return models.Model(inp, out)

model = build_residual_denoiser()
model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss="mse",
              metrics=[tf.keras.metrics.MeanAbsoluteError()])

model.summary()

# -----------------------------
# 6) Train
# -----------------------------
callbacks = [
    tf.keras.callbacks.ReduceLROnPlateau(patience=3, factor=0.5, verbose=1),
    tf.keras.callbacks.EarlyStopping(patience=6, restore_best_weights=True, verbose=1),
]

hist = model.fit(
    Xtr, Ytr,
    validation_data=(Xva, Yva),
    epochs=40,
    batch_size=32,
    callbacks=callbacks,
    verbose=1
)

# -----------------------------
# 7) Quick qualitative check (plot one window, one channel)
# -----------------------------
k = 0
pred = model.predict(Xva[:1], verbose=0)[0]  # (WIN,6)
gt   = Yva[0]
inp0 = Xva[0]

ch = 0  # channel 0 corresponds to V1
plt.figure()
plt.plot(inp0[:,ch], label="Input (Rva) norm")
plt.plot(gt[:,ch],   label="GT (Ra) norm")
plt.plot(pred[:,ch], label="Pred norm")
plt.legend()
plt.title("Q7: Ventricular removal (one window, V1)")
plt.show()
```
